```python
from Nqueens import Nqueens
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
def test_time(max):
    indexs = []
    times = []
    steps = []
    solutions = []

    for size in range(0, max+1):
    
        NQ = Nqueens(size, False)
        print(f"Size: {size} COMPLETE")
        size, step, time, solution = NQ.output()

        indexs.append(size)
        times.append(time)
        steps.append(step)
        solutions.append(solution)

    plt.figure(figsize=(10, 6))

    plt.plot(indexs, times, label='Time (s-perf_counter)', marker='o')
    plt.plot(indexs, steps, label='Steps', marker='x')
    plt.plot(indexs, solutions, label='Solutions', marker='s')

    plt.yscale('log')
    plt.xlabel('Size')
    plt.ylabel('Values (Log Scale)')
    plt.title('Performance Metrics of N-Queens')
    plt.legend(loc='best', fontsize=10)  
    plt.grid(True, which="both", linestyle='--', linewidth=0.5)
    plt.show()

    
```

```python
#PLOT THE RESULTS OF FUNCTION
runStats = test_time(15)
```
